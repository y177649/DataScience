# Speeding up scalar selection

```python
#college = pd.read_csv('data/college.csv', index_col='INSTNM')
college = pd.read_csv('https://takafumikubota.jp/pandas/data/college.csv', index_col='INSTNM')
cn = 'Texas A & M University-College Station'
college.loc[cn, 'UGDS_WHITE']
```

```python
college.at[cn, 'UGDS_WHITE']
```

```python
%timeit college.loc[cn, 'UGDS_WHITE']
```

```python
%timeit college.at[cn, 'UGDS_WHITE']
```

.atが.locよりも速い理由：


1.   単純性と特化：
.at[] は1つの要素のみを取得するように特化しているため、内部的には必要なチェックとオーバーヘッドが少なくて済みます。.loc[] は一度に複数の要素を選択できるように設計されているため、インデックスを解決するためのより複雑なロジックを実行する必要があります。
2.   オーバーヘッドの削減：
.at[] は単一の値に直接アクセスするための簡素化されたメソッドであり、より複雑なインデックス解決やエラーチェックが不要です。.loc[] では範囲や条件に基づく選択のための追加のオーバーヘッドが存在します。
3.   内部実装：
.at[] は内部的にデータの格納方法に密接にアクセスし、必要最小限のステップでデータを取得します。これに対して .loc[] はより一般的なパスを通じてデータにアクセスするため、若干のパフォーマンスコストが発生します。

したがって、単一のスカラー値を選択する場合は .at[] を使用すると、より効率的で速い結果が得られます。一方、複数の要素や複雑なインデックスを使用する場合は .loc[] が適していますが、これには多少のパフォーマンスコストが伴います。

```python
row_num = college.index.get_loc(cn)
col_num = college.columns.get_loc('UGDS_WHITE')
```

```python
row_num, col_num
```

```python
%timeit college.iloc[row_num, col_num]
```

```python
%timeit college.iat[row_num, col_num]
```

```python
%timeit college.iloc[5, col_num]
```

```python
%timeit college.iat[5, col_num]
```
