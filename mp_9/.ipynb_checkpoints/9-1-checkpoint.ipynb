```python
import csv
import pandas as pd
import numpy as np

drink = pd.read_csv('https://takafumikubota.jp/pandas/data/kakei_softdrink.csv',encoding='cp932')

data_ = drink[['pref', 'tea2021', 'tea2022', 'och2021', 'och2022']]

data_['zennen_tea'] = (data_['tea2022'] / data_['tea2021']) * 100
data_['zennen_och'] = (data_['och2022'] / data_['och2021']) * 100

data_['tea_status'] = data_['zennen_tea'] >= 100
data_['och_status'] = data_['zennen_och'] >= 100

tea_true = data_['tea_status'].sum()
och_true = data_['och_status'].sum()

print('Sum_perf = 47')
print(f'Tea True = {tea_true}')
print(f'Och True = {och_true}')

data_
```
